 R√©gression lin√©aire avec le n≈ìud de r√©gression Dans cet exercice, nous construisons un mod√®le de r√©gression pour le

1. Importez les donn√©es et sp√©cifiez les types de variables avec le n≈ìud Type. 

```python
dataf = pd.read_csv('Datasets/housing.data.txt' , sep=",")
# Affichage de la taille du dataset (n_lignes and n_colonnes)
print("housing's shape : ", dataf.shape)
# Affichage des 10 premi√®res lignes
dataf.head(11)
```

```python
dataf.isnull().sum()
```

2. Ajoutez un n≈ìud de r√©gression au flux et s√©lectionnez MEDV comme variable cible et toutes les autres variables comme variable contribution.

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd  
import seaborn as sns 
%matplotlib inline
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(dataf['MEDV'], bins=30)
plt.show()
```

```python
correlation_matrix = dataf.corr().round(2)
# annot = True to print the values inside the square
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
plt.figure(figsize=(20, 5))

features = ['LSTAT', 'RM']
target = dataf['MEDV']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = dataf[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

```python
X = pd.DataFrame(np.c_[dataf['LSTAT'], dataf['RM']], columns = ['LSTAT','RM'])
Y = dataf['MEDV']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score 

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
# model evaluation for training set
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print(" performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# model evaluation for testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
